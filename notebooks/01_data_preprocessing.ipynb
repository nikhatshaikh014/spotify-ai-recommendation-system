## Clean Genres

```python
import ast

def clean_genres(genre):
    try:
        # If it's a list-like string, convert it
        if genre.startswith("["):
            genre_list = ast.literal_eval(genre)
            # Replace spaces with underscore in each genre
            genre_list = [g.replace(" ", "_") for g in genre_list]
            return " ".join(genre_list)
        else:
            # Single genre case
            return genre.replace(" ", "_")
    except:
        return "unknown"

df["clean_genres"] = df["artist_genres"].apply(clean_genres)

df[["artist_genres", "clean_genres"]].head()
```

## Select Numeric Features

```python
numeric_features = [
    "track_popularity",
    "artist_popularity",
    "artist_followers",
    "track_duration_min"
]

df_numeric = df[numeric_features]
df_numeric.head()
```

## Scale Numeric Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_numeric = scaler.fit_transform(df_numeric)

scaled_numeric[:5]
```

## Vectorize Genres (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=1000)

genre_matrix = tfidf.fit_transform(df["clean_genres"])

genre_matrix.shape
```

## Combine Numeric + Genre Features

```python
from scipy.sparse import hstack

final_matrix = hstack([genre_matrix, scaled_numeric])

final_matrix.shape
```

Cosine Similarity

Why?

Because:

Works well with TF-IDF

Works well with high dimensional vectors

Scale independent

Standard for recommender systems

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(final_matrix)

similarity_matrix.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

pop_scaler = MinMaxScaler()

df["pop_score"] = pop_scaler.fit_transform(
    df[["track_popularity"]]
)

df[["track_name", "pop_score"]].head()
```

```python
def assign_mood(row):
    genres = row["clean_genres"]
    pop = row["track_popularity"]
    
    if any(g in genres for g in ["rap", "hip_hop", "synthpop", "pop"]) and pop > 60:
        return "Energetic"
    
    elif any(g in genres for g in ["indie", "soft_pop", "folk", "r&b"]):
        return "Chill"
    
    elif any(g in genres for g in ["metal", "dark", "soundtrack", "alternative"]):
        return "Emotional"
    
    elif pop > 80:
        return "Trending"
    
    else:
        return "Neutral"

df["mood"] = df.apply(assign_mood, axis=1)

df["mood"].value_counts()
```

## Create Recommendation Function

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def recommend(song_name, top_n=5, alpha=0.8, mood=None):
    
    idx = df[df["track_name"] == song_name].index[0]
    
    # Compute similarity
    sim_scores = cosine_similarity(
        final_matrix[idx], 
        final_matrix
    ).flatten()
    
    # Blend similarity + popularity
    blended_scores = alpha * sim_scores + (1 - alpha) * df["pop_score"]
    
    # Create temp dataframe
    temp_df = df.copy()
    temp_df["score"] = blended_scores
    
    # Filter by mood if provided
    if mood:
        temp_df = temp_df[temp_df["mood"] == mood]
    
    # Sort by score
    temp_df = temp_df.sort_values("score", ascending=False)
    
    # Remove itself
    temp_df = temp_df[temp_df["track_name"] != song_name]
    
    return temp_df[
        ["track_name", "artist_name", "mood", "track_popularity"]
    ].head(top_n)
```
